```
df=spark.read.csv("/FileStore/tables/iris.data",inferSchema='true')
display(df)
```

```
#importing
from pyspark.sql.functions import *
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.feature import StringIndexer
```

```
#renaming column
iris_df= df.select(col("_c0").alias("sepal_length"), col("_c1").alias("sepal_width"), col("_c2").alias("petal_length"), col("_c3").alias("petal_width"), col("_c4").alias("species"))
iris_df.show()
```

```
#creating features column
VectorAssembler2= VectorAssembler(inputCols=["sepal_length","sepal_width","petal_length","petal_width"], outputCol="features")
viris_df=VectorAssembler2.transform(iris_df)
viris_df.show()
```

```
#labeling
indexer2=StringIndexer(inputCol="species", outputCol="label")
iviris_df= indexer2.fit(viris_df).transform(viris_df)
iviris_df.show(2)
```

```
from pyspark.ml.classification import NaiveBayes
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```
splits= iviris_df.randomSplit([0.6,0.4],1)
train_df=splits[0]
test_df=splits[1]
```

```
train_df.count()
```

```
test_df.count()
```

```
#creating model
nb=NaiveBayes(modelType="multinomial")# we have three variables, so modelType="multinomial"
nbmodel=nb.fit(train_df)
```

```
#making prdictions
predictions_df=nbmodel.transform(test_df)
predictions_df.take(1)
```

```
#evaluating model
evaluator= MulticlassClassificationEvaluator(labelCol="label",predictionCol="prediction", metricName="accuracy") #we want accuracy, so metricName="accuracy"
nbaccuracy=evaluator.evaluate(predictions_df)
nbaccuracy
```

```
#the accuracy is not bad considering we used Naive Bayes classification.
```

```
from pyspark.ml.classification import MultilayerPerceptronClassifier
```

```
layersList=[4,5,6,3]
```

```
mlp= MultilayerPerceptronClassifier(layers=layersList, seed=1)
mlp_model= mlp.fit(train_df)
```

```
mlp_predictions= mlp_model.transform(test_df)
```

```
mlp_evaluator=MulticlassClassificationEvaluator(metricName="accuracy")
mlp_accuracy = mlp_evaluator.evaluate(mlp_predictions)
mlp_accuracy
```

```
#th e accuarcy is more than 94% indication Multilayer Perceptron Classification is better than Naive Bayes classification for iris data
```

```
#importing
from pyspark.ml.classification import DecisionTreeClassifier
```

```
#fitting
dt= DecisionTreeClassifier(labelCol="label", featuresCol="features")
dt_model= dt.fit(train_df)
```

```
#making predictions
dt_predictions= dt_model.transform(test_df)
```

```
#evaluation
dt_evaluator= MulticlassClassificationEvaluator(labelCol="label",predictionCol="prediction", metricName="accuracy")
```

```
#accuracy calculation
dt_accuracy= dt_evaluator.evaluate(dt_predictions)
dt_accuracy
```

```
#the accuracy is quite similar to indication Multilayer Perceptron Classification.
```
