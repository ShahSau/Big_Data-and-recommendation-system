```
df= spark.read.csv('/FileStore/tables/clustering_dataset.csv', header='true', inferSchema='true')
display(df)
```

```
df.printSchema()
```

```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.clustering import KMeans
```

```
#creating a feature column
VectorAssembler2 = VectorAssembler(inputCols=["col1","col2","col3"], outputCol="features")
#putting the feature column in the new data frame
new_df= VectorAssembler2.transform(df)
#showing the new df
new_df.show()
```

```
#seting up the KModels
km=  KMeans().setK(3)
km= km.setSeed(1)
kmodel= km.fit(new_df)
```

```
centers=kmodel.clusterCenters()
centers
```
